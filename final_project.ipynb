#Exploring the dataset

```python
import pandas as pd
import numpy as np

jobs_df = pd.read_csv('ds_salaries.csv')
```

```python
# see the first 5 rows
jobs_df.head()
```

```python
jobs_df = jobs_df.drop(columns=['Unnamed: 0'])
jobs_df.head()
```

```python
# used tali to see last 5 rows
jobs_df.tail()
```

```python
# length and shape of the dataset
print(len(jobs_df))
print(jobs_df.shape)
```

```python
jobs_df.info()
```

```python
# show some Statistics
jobs_df.describe()
```

```python
# chech if there is null value
jobs_df.isnull().sum()
```

```python
# extract column names
column_names = list(jobs_df.columns)
column_names
```

```python
# exploring unique values
for name in column_names:
  if name == 'salary' or name == 'salary_currency' or name == 'salary_in_usd':
    print()
  else:
    print(name)
    print(jobs_df[name].unique())
```

```python
# cheching how many time a value repeated
for name in column_names:
  if name == 'salary' or name == 'salary_currency' or name == 'salary_in_usd':
    print()
  else:
    print(name)
    print(jobs_df[name].value_counts())
```
